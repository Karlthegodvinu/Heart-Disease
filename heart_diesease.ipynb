```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import LabelEncoder
```

```python
heart.info()
```

```python
X = heart.drop(columns='Outcome')
Y = heart['Outcome']
```

```python
X
```

```python
Y
```

```python
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
```

```python
categorical_cols
```

```python
encoder = OneHotEncoder(sparse_output=False, drop='first')
```

```python
encoded_data = encoder.fit_transform(X[categorical_cols])
```

```python
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))
```

```python
X_encoded = pd.concat([X.drop(columns=categorical_cols), encoded_df], axis=1)
```

```python
print(X_encoded)
```

```python
X_encoded
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_normalized = pd.DataFrame(scaler.fit_transform(X_encoded), columns=X_encoded.columns)
```

```python
le = LabelEncoder()
Y_encoded = le.fit_transform(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_normalized, Y_encoded, test_size=0.2, random_state=42, stratify=Y_encoded)
```

```python
model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
label_mapping = {0: "No Heart Attack", 1: "Heart Attack"}
Y_pred_labels = np.vectorize(label_mapping.get)(Y_pred)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy:.4f}")
print("Classification Report:\n", classification_report(Y_test, Y_pred))
```

```python
def predict_heart_attack(user_input):
    user_df = pd.DataFrame([user_input], columns=X_normalized.columns)

    user_df_scaled = scaler.transform(user_df)

    prediction = model.predict(user_df_scaled)[0]

    return "Heart Attack" if prediction == 1 else "No Heart Attack"

user_input_example = X_normalized.iloc[0].to_dict()

result = predict_heart_attack(user_input_example)
print("Prediction:", result)
```
